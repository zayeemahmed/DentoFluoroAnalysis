```python
import tensorflow as tf

import numpy as np

import keras

import os

from keras.src.applications.imagenet_utils import preprocess_input

from keras.layers import Dense,Flatten,Conv2D,Activation,Dropout,DepthwiseConv2D

from keras import backend as K

from keras.models import Sequential, Model

from keras.models import load_model

from keras.optimizers import Adam

from keras.callbacks import EarlyStopping,ModelCheckpoint

from keras.layers import MaxPool2D, GlobalAveragePooling2D,Rescaling,AveragePooling2D

import matplotlib.pyplot as plt

from keras.regularizers import l2

from keras.preprocessing.image import ImageDataGenerator
```

```python
train_dataset = ImageDataGenerator(rescale=1/255).flow_from_directory(batch_size=25,directory='/content/drive/MyDrive/DataSet2/Train',shuffle=True,target_size=(224,224),class_mode='categorical')

validation_dataset = ImageDataGenerator(rescale=1/255).flow_from_directory(batch_size=10,directory='/content/drive/MyDrive/DataSet2/Valid',shuffle=True,target_size=(224,224),class_mode='categorical')
```

```python

m=keras.Sequential()

m.add(Conv2D(filters=64,padding='same',strides=2,kernel_size=3,activation='relu',input_shape=(224,224,3)))
m.add(MaxPool2D(pool_size=2,strides=2))

m.add(Conv2D(filters=32,padding='same',strides=2,kernel_size=3,activation='relu'))
m.add(MaxPool2D(pool_size=2,strides=2))

m.add(Flatten())
#m.add(Dense(512,activation='relu'))
#m.add(Dense(256,activation='relu'))
m.add(Dense(128,activation='relu'))
m.add(Dense(64,activation='relu'))
m.add(Dropout(0.5))
m.add(Dense(4,activation='softmax'))
model=Model(inputs=m.input,outputs=m.output)
```

```python
model.summary()
```

```python
#Model Fit and Compile
modelPath = '/content/drive/MyDrive/best-motebooks'
if not os.path.exists(modelPath):
  os.makedirs(modelPath)
  print('Model Directory Created')
else:
  print('Model Directory Already Exists')

model_checkpoint = tf.keras.callbacks.ModelCheckpoint(modelPath+'/best-model2.h5', monitor='val_accuracy',
                                                      verbose=1, save_best_only=True, mode='auto')
optimizer=Adam(learning_rate=0.001)
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
History = model.fit(train_dataset, validation_data=validation_dataset, epochs=40,callbacks=model_checkpoint)
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
loss_train = History.history['accuracy']
loss_val = History.history['val_accuracy']
epochs = range(1,41)
plt.plot(epochs, loss_train, 'g', label='Training accuracy')
plt.plot(epochs, loss_val, 'b', label='validation accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
loss_train = History.history['loss']
loss_val = History.history['val_loss']
epochs = range(1,41)
plt.plot(epochs, loss_train, 'g', label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
# Evaluate the Best Saved Model
bestModel = tf.keras.models.load_model('/content/drive/MyDrive/best-motebooks/best-model2.h5')
loss, accuracy= bestModel.evaluate(validation_dataset, batch_size=30, verbose=1)
print('Model Accuracy: {:0.2f} | Model Loss: {:0.4f}'.format(accuracy,loss))
```
